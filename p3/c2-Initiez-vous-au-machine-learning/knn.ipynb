```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml(name="mnist_784", version=1)
```

```python
mnist.data.values
```

```python
import numpy as np
```

```python
sample = np.random.randint(mnist.data.shape[0], size=5000)
data_sample = mnist.data.values[sample]
target_sample = mnist.target.values[sample]
```

```python
from sklearn.utils import resample
data_sample, target_sample = resample(mnist.data.values, mnist.target.values, n_samples=5000, random_state=0)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(data_sample, target_sample, train_size=0.8, random_state=0)
```

```python
ytest.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

errors = []
best_knn = KNeighborsClassifier()
for k in range(2, 15):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(xtrain, ytrain)
    error = 1 - knn.score(xtest, ytest)
    errors.append(error)
    if np.array(errors).min() == error:
        best_knn = knn
    
errors
```

```python
plt.plot(range(2,15), np.array(errors)*100, 'o-')
```

```python
best_knn.n_neighbors
```

```python
predict = best_knn.predict(xtest)
```

```python
n_samples = [5,5]

xdisplay, ydisplay, predictdisplay = resample(xtest, ytest, predict, n_samples=n_samples[0]*n_samples[1], random_state=0)

fig, ax = plt.subplots(n_samples[0], n_samples[1])

for index, value in enumerate(xdisplay):
    plt.subplot(n_samples[0], n_samples[1], index + 1)
    number = np.reshape(xdisplay[index], (28,28))
    plt.imshow(number, cmap="Greys", interpolation="nearest")
    plt.title(f"Pred={predictdisplay[index]}", y=-0.2)
    plt.axis('off')

plt.show()
```
