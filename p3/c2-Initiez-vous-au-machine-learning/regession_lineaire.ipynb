```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
house_data = pd.read_csv('house.csv')
house_data.head()
```

```python
plt.scatter(house_data['surface'], house_data['loyer'], s=4)
plt.show()
```

```python
house_data = house_data.loc[house_data['loyer'] < 10000, :]
plt.scatter(house_data['surface'], house_data['loyer'], s=4)
plt.show()
```

```python
X = np.array(house_data['surface']).reshape(-1, 1)
X[:10]
```

```python
y = np.array(house_data['loyer']).reshape(-1, 1)
y[:10]
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()

model.fit(X, y)
```

```python
model.intercept_
```

```python
plt.plot([0, 250], [model.intercept_, 250*model.coef_ + model.intercept_], linestyle='--')
plt.scatter(house_data['surface'], house_data['loyer'], s=4)
plt.show()
```

```python
model.predict([[150]])
```
