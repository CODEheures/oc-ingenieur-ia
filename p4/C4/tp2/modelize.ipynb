```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import timeit

def train_and_predict(X_train, y_train, X_test, y_test, n_estimators):
    start = timeit.default_timer()

    rfc = RandomForestClassifier(n_estimators=n_estimators, n_jobs=-1)
    rfc.fit(X_train, y_train)
    end_train = timeit.default_timer()

    pred = rfc.predict(X_test)
    end_predict = timeit.default_timer()

    accuracy = accuracy_score(y_test, pred)

    time_train = end_train-start
    time_predict = end_predict-end_train


    print(f'accuracy: {accuracy:.2f}, time train: {time_train:.2f}, time predict: {time_predict:.2f}')
    
    return rfc
```

```python
rfc = train_and_predict(X_train, y_train, X_test, y_test, 1000)
```

```python
from sklearn.feature_selection import SelectFromModel

select = SelectFromModel(rfc, prefit=True, threshold=0.002)
X_train2 = select.transform(X_train)
X_test2 = select.transform(X_test)

X_train2.shape
```

```python
rfc2 = train_and_predict(X_train2, y_train, X_test2, y_test, 250)
```
