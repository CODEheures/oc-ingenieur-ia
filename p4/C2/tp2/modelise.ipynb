```python
import pandas as pd
```

```python
df = pd.read_csv('./data/clean/cleaned.csv')
df.head()
```

```python
X = df.drop(columns=['target'])
y = df['target']
```

```python
X
```

```python
y
```

```python
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

```python
from sklearn import dummy
estimator = dummy.DummyClassifier(strategy='most_frequent')
estimator.fit(X_train, y_train)
```

```python
def score(estimator):
    """ Imprime les scores accuracy sur train et test pour un estimator"""
    
    score_test = estimator.score(X_test, y_test).round(4)
    score_train = estimator.score(X_train, y_train).round(4)
    print(f'Score train: {score_train}')
    print(f'Score test: {score_test}')
```

```python
score(estimator)
```

```python
from sklearn import metrics
def confusion(y_true, y_pred):
    matrix = metrics.confusion_matrix(y_true, y_pred)
    matrix = pd.DataFrame(matrix)
    matrix.rename(columns={0: 'predit comestible', 1: 'predit toxique'}, index={0: 'comestible', 1: 'toxique'}, inplace=True)
    return matrix
```

```python
confusion(y_test, estimator.predict(X_test))
```

```python
from sklearn import linear_model
estimator = linear_model.LogisticRegression(solver='liblinear')
estimator.fit(X_train, y_train)

score(estimator)
confusion(y_test, estimator.predict(X_test))
```

```python
estimator.predict_proba(X_test).round(2)
```

```python
from sklearn import svm
estimator = svm.LinearSVC()
estimator.fit(X_train, y_train)
score(estimator)
confusion(y_test, estimator.predict(X_test))
```

```python
import numpy as np
estimator = linear_model.LogisticRegression()
params = {'C': np.logspace(-3,3,7),
          'penalty': ['l1', 'l2'],
          'solver': ['lbfgs', 'liblinear', 'newton-cg']
         }

grid = model_selection.GridSearchCV(estimator=estimator,cv=10,n_jobs=-1,param_grid=params,return_train_score=True,verbose=True)

grid.fit(X_train, y_train)
```

```python
best_params = grid.best_params_
best_params
```

```python
def grid_result(grid):
    res = grid.cv_results_
    res = pd.DataFrame(res)
    cols = [col for col in res.columns if 'split' not in col]
    res = res[cols].sort_values('mean_test_score', ascending=False)
    return res.head()
```

```python
grid_result(grid)
```

```python
estimator = linear_model.LogisticRegression(**grid.best_params_)
estimator.fit(X_train, y_train)
score(estimator)
confusion(y_test, estimator.predict(X_test))
```

```python
estimator = svm.LinearSVC()
params = {'C': np.logspace(-3,3,7), 'penalty': ['l1', 'l2'],
          'dual': [True, False]}

grid = model_selection.GridSearchCV(estimator=estimator,cv=10,n_jobs=-1,param_grid=params,return_train_score=True,verbose=True)

grid.fit(X_train, y_train)
```

```python
grid_result(grid)
```

```python
estimator = svm.LinearSVC(**grid.best_params_)
estimator.fit(X_train, y_train)
score(estimator)
confusion(y_test, estimator.predict(X_test))
```
