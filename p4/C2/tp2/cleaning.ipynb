```python
import pandas as pd
```

```python
raw_data = pd.read_csv('./data/source/mushrooms.csv')
raw_data.head()
```

```python
raw_data.info()
```

```python
raw_data.describe()
```

```python
raw_data.nunique()
```

```python
raw_data.rename(columns={"class": "target"}, inplace=True)
raw_data.head()
```

```python
tmp = raw_data.nunique()
tmp
```

```python
tmp[tmp==1]
```

```python
columns = tmp[tmp==1].index
columns
```

```python
raw_data.drop(columns, axis=1, inplace=True)
raw_data.nunique()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for c in raw_data.columns:
    raw_data[c] = le.fit_transform(raw_data[c])
```

```python
raw_data.head()
```

```python
raw_data.to_csv('./data/clean/cleaned.csv', index=False)
```
