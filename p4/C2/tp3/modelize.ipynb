```python
import pandas as pd

data = pd.read_csv('./data/Dataset_feuilles_1.csv')
data.head()
```

```python
test = pd.read_csv('./data/test.csv')
test.head()
```

```python
X = data.iloc[:, 2:]
y = data.iloc[:,1]
```

```python
y
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
pd.DataFrame(X_scaled).describe().round(2)
```

```python
encoder = preprocessing.LabelEncoder()
y = encoder.fit_transform(y)
```

```python
from sklearn import model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
from sklearn import neighbors

estimator = neighbors.KNeighborsClassifier(n_jobs=-1)

estimator.fit(X_train, y_train)
```

```python
import numpy as np
baseline_accuracy = estimator.score(X=X_test, y=y_test)
baseline_accuracy
```

```python
from sklearn import multiclass
from sklearn.svm import LinearSVC
from sklearn import model_selection

estimator = multiclass.OneVsOneClassifier(n_jobs=-1, estimator=LinearSVC(random_state=40))

params = {'estimator__penalty': ['l1','l2'], 'estimator__C': np.logspace(-3,3,7)}

grid = model_selection.GridSearchCV(estimator=estimator, param_grid=params, n_jobs=-1, cv=3, return_train_score=True, verbose=True, scoring='accuracy')

grid.fit(X_train, y_train)

best_score = grid.best_score_
best_score
```

```python
grid.best_params_
```

```python
estimator = multiclass.OneVsOneClassifier(n_jobs=-1, estimator=LinearSVC(penalty='l2', C=0.01))

estimator.fit(X_train, y_train)
score = estimator.score(X_test, y_test)
score
```

```python
estimator = multiclass.OneVsRestClassifier(n_jobs=-1, estimator=LinearSVC(random_state=40))

params = {'estimator__penalty': ['l1','l2'], 'estimator__C': np.logspace(-3,3,7)}

grid = model_selection.GridSearchCV(estimator=estimator, param_grid=params, n_jobs=-1, cv=3, return_train_score=True, verbose=True, scoring='accuracy')

grid.fit(X_train, y_train)

best_score = grid.best_score_
best_score
```

```python
grid.best_params_
```

```python
estimator = multiclass.OneVsRestClassifier(n_jobs=-1, estimator=LinearSVC(penalty='l2', C=0.01, random_state=40))

estimator.fit(X_train, y_train)
score = estimator.score(X_test, y_test)
score
```
