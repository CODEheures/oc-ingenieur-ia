```python
commandes = pd.read_csv('commande.csv')
produits = pd.read_csv('produits.csv')
produits.head()
```

```python
commandes.head()
```

```python
conso = pd.merge(commandes, produits, on='id', how='left')
conso['CA'] = conso['nombre'] * conso['prix']
conso.head()
```

```python
consog = conso.groupby('id').agg({'nombre': 'sum', 'CA': 'sum'})
```

```python
consog.head()
```

```python
consog.sort_values('CA')
```
